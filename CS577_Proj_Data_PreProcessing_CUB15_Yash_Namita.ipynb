```
def load_class_ids(class_info_file_path):  
    with open(class_info_file_path, 'rb') as f:
        class_ids = pickle.load(f, encoding='latin1')
    return(class_ids) 

def load_filenames(filenames_file_path):
    with open(filenames_file_path, 'rb') as f:
        filenames = pickle.load(f, encoding='latin1')
    return(filenames)

def load_embeddings(embeddings_file_path):
    with open(embeddings_file_path, 'rb') as f:
        embeddings = pickle.load(f, encoding='latin1')
        embeddings = np.array(embeddings)
    return(embeddings)
    
```

```
def merge_train_list(class_info_file_path, filenames_file_path, embeddings_file_path):
  class_ids = load_class_ids(class_info_file_path)
  #print(class_ids[0])
  #print(len(class_ids))

  filenames = load_filenames(filenames_file_path)
  #print(filenames[0])
  #print(len(filenames))

  embeddings = load_embeddings(embeddings_file_path)
  print(embeddings.shape)
  #print(embeddings[0])

  merged_train_list = []
  temp_list=[]

  for i in range(len(class_ids)):
    temp_list = [class_ids[i] , filenames[i] ,embeddings[i]]
    merged_train_list.append(temp_list)
    #print(merged_list)

  #print(merged_train_list)
  #print(len(merged_train_list))
  return merged_train_list


#Function Call
#merge_train_list("./train/class_info.pickle", "./train/filenames.pickle", "./train/char-CNN-RNN-embeddings.pickle")
```

```
#Now combine the data from test and train into a single combined dataset
#For our project we will be just focusing on filenames and class_ids

def load_dataset(image_pickle_file_path,filenames_file_path, class_info_file_path, cub_dataset_dir, image_size):
    """
    Load dataset
    """
    image_pickle_file = image_pickle_file_path
    filenames = load_filenames(filenames_file_path)
    class_ids = load_class_ids(class_info_file_path)
    
    with open(image_pickle_file,'rb') as f:
      images =pickle.load(f , encoding='latin1')

    X, y= [], []

    print("Load dataset for loop 1 starts")
    for index, filename in enumerate(filenames):
      #bounding_box = bounding_boxes[filename]
      try:
        y.append(class_ids[index])
      except Exception as e:
        print(e)
    print("Load dataset for loop 1 completed")
    X = np.array(images)
    y = np.array(y)
    return X, y
```

```
augmented_combined_dataset = []

for i in range(len(combined_dataset)):
  image = combined_dataset[i][0].reshape(-1,64,64,3)  
  aug_iter = X_datagen.flow(image)
   
  aug_images = [next(aug_iter)[0].astype(np.uint8) for j in range(10)]
  augmented_combined_dataset.append([aug_images, combined_dataset[i][1]])
```
